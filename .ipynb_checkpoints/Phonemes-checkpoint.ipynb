```python
phone = {}

for word in newdict:
    try:
        phone.update({word:arpabet[word][0]})
    except:
        try:
            counter = 0
            for i in word:
                substring = word[0:1+counter]
                counter += 1
                try:
                    print(substring, arpabet[substring][0])
                except Exception as e:
                    print(e)
        except Exception as e:
            print(e)
```

```python
import pyphen
dic = pyphen.Pyphen(lang='en')
```

```python
for word in newdict:
    print(dic.inserted(word))
```

```python

d = cmudict.dict()
missing_words = []
for word in newdict:
    try:
        print(d[word])
        print(nltk.pos_tag(nltk.word_tokenize(word)))
        print(dic.inserted(word))
        if '-' in dic.inserted(word):
            syllables = dic.inserted(word).split('-')
            for syl in syllables:
                print(syl)
        print("---")
    except:
        print("Missing word: ", word)
        missing_words.append(word)
    
print(missing_words)
```
